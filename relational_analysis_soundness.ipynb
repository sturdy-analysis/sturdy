```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("relational-test-script.csv")

control_flow = df[
    (df['filename'] == 'block.wast') |
    (df['filename'] == 'br.wast') |
    (df['filename'] == 'br_if.wast') |
    (df['filename'] == 'br_table.wast') |
    (df['filename'] == 'if.wast') |
    (df['filename'] == 'labels.wast') |
    (df['filename'] == 'left-to-right.wast') |
    (df['filename'] == 'nop.wast') |
    (df['filename'] == 'return.wast') |
    (df['filename'] == 'switch.wast') |
    (df['filename'] == 'unreachable.wast') |
    (df['filename'] == 'unreached-invalid_wasm2_compatible.wast')
]
control_flow.insert(1, 'category', 'Control Flow')

conversions = df[
    (df['filename'] == 'conversions.wast')
]
conversions.insert(1, 'category', 'Conversions')

floats = df[
    (df['filename'] == 'const.wast') |
    (df['filename'] == 'f32.wast') |
    (df['filename'] == 'f32_bitwise.wast') |
    (df['filename'] == 'f32_cmp.wast') |
    (df['filename'] == 'f64.wast') |
    (df['filename'] == 'f64_bitwise.wast') |
    (df['filename'] == 'f64_cmp.wast') |
    (df['filename'] == 'float_exprs.wast') |
    (df['filename'] == 'float_literals.wast') |
    (df['filename'] == 'float_memory.wast') |
    (df['filename'] == 'float_misc.wast')
]
floats.insert(1, 'category', 'Floats')

functions = df[
    (df['filename'] == 'call.wast') |
    (df['filename'] == 'call_indirect.wast') |
    (df['filename'] == 'elem.wast') |
    (df['filename'] == 'fac.wast') |
    (df['filename'] == 'forward.wast') |
    (df['filename'] == 'func.wast') |
    (df['filename'] == 'func_ptrs.wast') |
    (df['filename'] == 'skip-stack-guard-page.wast') |
    (df['filename'] == 'stack.wast') |
    (df['filename'] == 'table_wasm2_compatible.wast')
]
functions.insert(1, 'category', 'Functions')

global_vars = df[
    (df['filename'] == 'global.wast')
]
global_vars.insert(1, 'category', 'Global Vars.')

integers = df[
    (df['filename'] == 'i32.wast') |
    (df['filename'] == 'i64.wast') |
    (df['filename'] == 'int_exprs.wast') |
    (df['filename'] == 'int_literals.wast') |
    (df['filename'] == 'traps.wast')
]
integers.insert(1, 'category', 'Integers')

local_vars = df[
    (df['filename'] == 'local_get.wast') |
    (df['filename'] == 'local_set.wast') |
    (df['filename'] == 'local_tee.wast')
]
local_vars.insert(1, 'category', 'Local Vars.')

memory = df[
    (df['filename'] == 'address.wast') |
    (df['filename'] == 'align.wast') |
    (df['filename'] == 'data.wast') | 
    (df['filename'] == 'endianness.wast') |
    (df['filename'] == 'load.wast') | 
    (df['filename'] == 'store.wast') | 
    (df['filename'] == 'memory.wast') | 
    (df['filename'] == 'memory_grow.wast') | 
    (df['filename'] == 'memory_redundancy.wast') | 
    (df['filename'] == 'memory_size.wast') | 
    (df['filename'] == 'memory_trap.wast')
]
memory.insert(1, 'category', 'Memory')

modules = df[
    (df['filename'] == 'binary-leb128.wast') |
    (df['filename'] == 'comments.wast') |
    (df['filename'] == 'custom.wast') |
    (df['filename'] == 'exports.wast') |
    (df['filename'] == 'imports_wasm2_compatible.wast') |
    (df['filename'] == 'inline-module.wast') |
    (df['filename'] == 'linking.wast') |
    (df['filename'] == 'names.wast') |
    (df['filename'] == 'start.wast') |
    (df['filename'] == 'token.wast') |
    (df['filename'] == 'type.wast') |
    (df['filename'] == 'utf8-custom-section-id.wast') |
    (df['filename'] == 'utf8-import-field.wast') |
    (df['filename'] == 'utf8-import-module.wast') |
    (df['filename'] == 'utf8-invalid-encoding.wast')
]
modules.insert(1, 'category', 'Modules')

stack = df[
    (df['filename'] == 'select.wast') |
    (df['filename'] == 'unwind.wast')
]
stack.insert(1, 'category', 'Stack')

data = pd.concat([control_flow, conversions, floats, functions, global_vars, integers, local_vars, memory, modules, stack])

display(data)

test_files = data[data['abstract_domain'] == 'Box'].groupby(by=['category'])['filename'].size()
assertions = data[data['abstract_domain'] == 'Box'].groupby(by=['category'])['test_cases'].sum()
polyhedra = data[data['abstract_domain'] == 'Polka'].groupby(by=['category'])['passed_cases'].sum()
octagon = data[data['abstract_domain'] == 'Octagon'].groupby(by=['category'])['passed_cases'].sum()
intervals = data[data['abstract_domain'] == 'Box'].groupby(by=['category'])['passed_cases'].sum()

pd.options.display.float_format = '{:,.2f}%'.format
display(pd.DataFrame({
    'Assertions': assertions,
    'Polyhedra': polyhedra / assertions * 100,
    'Octagon': octagon / assertions * 100,
    'Intervals': intervals / assertions * 100
}))

display(pd.DataFrame({
    'Assertions': [assertions.sum()],
    'Polyhedra': [polyhedra.sum() / assertions.sum() * 100],
    'Octagon': [octagon.sum() / assertions.sum() * 100],
    'Intervals': [intervals.sum() / assertions.sum() * 100]
}))


display(assertions.sum() - octagon.sum())
display((assertions.sum() - octagon.sum()) / assertions.sum() * 100)
```
