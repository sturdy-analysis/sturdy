```python
import pandas

pandas.options.mode.copy_on_write = True
pandas.options.display.float_format = '{:,.1f}'.format

svbench = pandas.read_csv('../relational-svbench.csv')
svbench.set_index(['suite','test_case'], inplace=True)

single = svbench[svbench['abstract_domain'] == 'Box']

result = pandas.DataFrame({
    # 'benchmark': single.index.get_level_values(1),
    'tests': single['abstract_domain'].groupby('suite').count(),
    # 'no_verdict': single['expected_verdict'].isna().groupby('suite').sum(),
    'with_verdict': single['expected_verdict'].notna().groupby('suite').count(),
    'octagon_passed': svbench[(svbench['abstract_domain'] == 'Octagon') & (svbench['passed'] == True)]['passed'].groupby('suite').count(),
    'interval_passed': svbench[(svbench['abstract_domain'] == 'Box') & (svbench['passed'] == True)]['passed'].groupby('suite').count(),
    'verdict_true': single[single['expected_verdict'] == True]['expected_verdict'].groupby('suite').count(),
    'octagon_verdict_true_passed': svbench[(svbench['abstract_domain'] == 'Octagon') & (svbench['expected_verdict'] == True) & (svbench['passed'] == True)]['passed'].groupby('suite').count(),
    'interval_verdict_true_passed': svbench[(svbench['abstract_domain'] == 'Box') & (svbench['expected_verdict'] == True) & (svbench['passed'] == True)]['passed'].groupby('suite').count(),
    'octagon_timeout': svbench[(svbench['abstract_domain'] == 'Octagon') & (svbench['timed_out'] == True)]['timed_out'].groupby('suite').count(),
    'interval_timeout': svbench[(svbench['abstract_domain'] == 'Box') & (svbench['timed_out'] == True)]['timed_out'].groupby('suite').count(),
})
result = result.fillna({'verdict_true': 0, 'octagon_verdict_true_passed': 0, 'interval_verdict_true_passed': 0, 'octagon_timeout': 0, 'interval_timeout': 0}).astype({'verdict_true': 'int64', 'octagon_verdict_true_passed': 'int64', 'interval_verdict_true_passed': 'int64', 'octagon_timeout': 'int64', 'interval_timeout': 'int64'})
display(result)
display(result['verdict_true'].sum())

result.to_csv('../relational-svbench-result.csv')

```
